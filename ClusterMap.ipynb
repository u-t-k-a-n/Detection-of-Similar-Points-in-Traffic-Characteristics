```python
import pandas as pd
import plotly.express as px
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = "drive/MyDrive/SimilarityScores/Coordinates/2017/ClusterCoordinates_2017.csv"
```

```python
df = pd.read_csv(path)
df = df.drop("Unnamed: 0",axis=1)
df
```

```python
df["cluster"] = df["cluster"].astype(str)
```

```python
fig = px.scatter_mapbox(df, lon='latitude', lat='longitude', hover_name='vSegID', hover_data=['vSegDir',"cluster"],
                        color="cluster", zoom=4, height=600)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
