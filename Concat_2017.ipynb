```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
import os
path = "drive/MyDrive/2017_v1"
files = list()
for dirname, _, filenames in os.walk(path):
    for filename in filenames:
        files.append(os.path.join(dirname, filename))
```

```python
df = pd.read_csv(files[0],index_col=["time"])
```

```python
df.drop(columns=["is_under_10","is_over_130"],inplace=True)
```

```python
df = df.rename(columns={"speed":files[0]})
```

```python
df
```

```python
for i in range(1,len(files)):
  df2 = pd.read_csv(files[i],index_col=["time"])
  df2.drop(columns=["is_under_10","is_over_130"],inplace=True)
  df2 = df2.rename(columns={"speed":files[i]})
  df = pd.concat([df,df2],axis=1)
```

```python
df
```

```python
df.to_csv("drive/MyDrive/2017_concat/Concat_2017.csv")
```
