```python
import os
import pandas as pd
import plotly.express as px
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path = "drive/MyDrive/Metadata/Coordinates_2017.csv"
```

```python
df = pd.read_csv(path)
```

```python
df
```

```python
fig = px.scatter_mapbox(df, lon='latitude', lat='longitude', hover_name='vSegID', hover_data=['vSegDir'],
                        color_discrete_sequence=["blue"], zoom=4, height=600)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
