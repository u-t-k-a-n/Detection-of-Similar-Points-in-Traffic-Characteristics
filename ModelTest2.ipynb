```python
def calc_error_rate(seg_id,seg_dir,dates):
  col_name = "drive/MyDrive/2017_v1/{}_{}_2017.csv".format(seg_id,seg_dir)
  date1 = dates[0]
  date2 = dates[1]
  date3 = dates[2]
  date4 = dates[3]
  date5 = dates[4]
  speed_predict = 0
  
  speed_predict += df2[df2["time"] == date1][col_name].iloc[0]
  speed_predict += df2[df2["time"] == date2][col_name].iloc[0]
  speed_predict += df2[df2["time"] == date3][col_name].iloc[0]
  speed_predict += df2[df2["time"] == date4][col_name].iloc[0]
  speed_predict /= 4
  speed_real = df2[df2["time"] == date5][col_name].iloc[0]

  return abs(speed_real - speed_predict) / speed_real
```
